```python
def generate(color, i):

    if i == len(color):
        yield color.copy()
        return
    for c in rgb :
        color[i] = c
        yield from generate(color, i+1)
```

```python
def test(color, relation):
    for i, rel in enumerate(relation):
        for j, r in enumerate(rel) :
            if color[i] == color[r] :
                return False
    return True
```
