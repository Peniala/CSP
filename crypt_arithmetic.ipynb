```python
import numpy as np
```

```python
def get_id(elmnt, letters):
    idx = [int(letters.index(e)) for e in elmnt]
    return idx

def generate(value, i):
    if i == len(value):
        yield value.copy()
        return
    
    for n in range(10):
        value[i] = n
        yield from generate(value, i+1)

def transform(idx, value):
    n = [int(value[x]) for x in idx]
    return int("".join(map(str, n)))

def test(value, l1_id, l2_id, res_id):
    value1 = transform(l1_id, value)
    value2 = transform(l2_id, value)
    res = transform(res_id, value)
    return value1 + value2 == res \
    and len(str(value1)) == len(l1_id) \
    and len(str(value2)) == len(l2_id) \
    and len(str(res)) == len(res_id) \
    and len(value) == len(set(value))
```

```python
l1 = "send"
l2 = "more"
res = "money"

letters = "".join(dict.fromkeys(l1 + l2 + res))

l1_id = get_id(l1, letters)
l2_id = get_id(l2, letters)
res_id = get_id(res, letters)

value = np.zeros(len(letters), dtype=int)
value = generate(value, 0)

for v in value:
    if test(v, l1_id, l2_id, res_id):
        print("Solution trouvée :", v)
        break
```
